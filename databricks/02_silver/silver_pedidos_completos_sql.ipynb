```python
%sql
-- ============================================================
-- CAMADA: SILVER
-- ENTIDADE: Pedidos Completos
--
-- OBJETIVO:
-- Criar uma visão consolidada dos pedidos, unificando
-- informações de clientes, pagamentos, itens e avaliações,
-- mantendo a granularidade de um pedido por linha e sem
-- aplicação de regras de negócio, facilitando o consumo
-- pelas camadas Gold, BI e modelos de Machine Learning.
--
-- FONTE:
-- workspace.olist_silver.pedidos_silver
-- workspace.olist_silver.pagamentos_agregados_silver
-- workspace.olist_silver.itens_agregados_pedido_silver
-- workspace.olist_silver.avaliacoes_por_pedido_silver
--
-- DESTINO:
-- workspace.olist_silver.pedidos_completos_silver
-- ============================================================

CREATE SCHEMA IF NOT EXISTS workspace.olist_silver;

CREATE OR REPLACE VIEW workspace.olist_silver.pedidos_completos_silver AS

SELECT
    p.order_id,
    p.customer_id,
    p.order_status,
    p.order_purchase_timestamp,
    p.order_delivered_customer_date,

    pay.total_payment_value,
    pay.max_installments,
    pay.payment_types_count,

    it.total_items,
    it.total_price,
    it.total_freight,
    it.distinct_product_categories,

    r.review_score

FROM workspace.olist_silver.pedidos_silver p

LEFT JOIN workspace.olist_silver.pagamentos_agregados_silver pay
       ON p.order_id = pay.order_id
LEFT JOIN workspace.olist_silver.itens_agregados_pedido_silver it
       ON p.order_id = it.order_id
LEFT JOIN workspace.olist_silver.avaliacoes_por_pedido_silver r
       ON p.order_id = r.order_id;
```
