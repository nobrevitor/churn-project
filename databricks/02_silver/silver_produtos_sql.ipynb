## Tabela Produtos:

### Objetivo:

Corrigir os erros identificados na camada bronze e criar a view produtos_silver a partir da camada Bronze. 

### Fonte:

`workspace.olist_bronze.produtos_bronze`

### Destino:

`workspace.olist_silver.produtos_silver`

```python
%sql
-- Garantindo a exitência do SCHEMA
CREATE SCHEMA IF NOT EXISTS workspace.olist_silver;
```

```python
%sql
-- Verificando a existência da tabela:
SELECT * FROM workspace.olist_bronze.produtos_bronze;
```

```python
%sql
-- Verificando quantas linhas tem na tabela:
SELECT COUNT(*) AS LINHAS
FROM workspace.olist_bronze.produtos_bronze;
```

Como os valores nulos das colunas `product_category_name`, `product_name_lenght`, `product_description_lenght` e `product_photos_qty` ocorrem nas mesmas linhas e representam menos de 2% do total dos registros, esses dados não agregam valor às análises e, portanto, essas linhas serão excluídas.

Também, as colunas classificadas como `BIGINT` serão tratadas para `INT`.

```python
%sql
SELECT 
      product_id,
      product_category_name,
      CAST(product_name_lenght AS INT) AS product_name_lenght,
      CAST(product_description_lenght AS INT) AS product_description_lenght,
      CAST(product_photos_qty AS INT) AS product_photos_qty,
      CAST(product_weight_g AS INT) AS product_weight_g,
      CAST(product_length_cm AS INT) AS product_length_cm,
      CAST(product_height_cm AS INT) AS product_height_cm,
      CAST(product_width_cm AS INT) AS product_width_cm
FROM workspace.olist_bronze.produtos_bronze
WHERE product_category_name IS NOT NULL
```

```python
%sql
CREATE OR REPLACE VIEW workspace.olist_silver.produtos_silver AS
SELECT 
      product_id,
      product_category_name,
      CAST(product_name_lenght AS INT) AS product_name_lenght,
      CAST(product_description_lenght AS INT) AS product_description_lenght,
      CAST(product_photos_qty AS INT) AS product_photos_qty,
      CAST(product_weight_g AS INT) AS product_weight_g,
      CAST(product_length_cm AS INT) AS product_length_cm,
      CAST(product_height_cm AS INT) AS product_height_cm,
      CAST(product_width_cm AS INT) AS product_width_cm
FROM workspace.olist_bronze.produtos_bronze
WHERE product_category_name IS NOT NULL
```
