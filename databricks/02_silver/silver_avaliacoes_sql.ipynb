## Tabela Avaliações:

### Objetivo:

Corrigir os erros identificados na camada bronze e criar a view avaliacoes_silver a partir da camada Bronze. 

### Fonte:

`workspace.olist_bronze.avaliacoes_bronze`

### Destino:

`workspace.olist_silver.avaliacoes_silver`

```python
%sql
-- Garantindo a exitência do SCHEMA
CREATE SCHEMA IF NOT EXISTS workspace.olist_silver;
```

```python
%sql
-- Verificando a existência da tabela:
SELECT * FROM workspace.olist_bronze.avaliacoes_bronze;
```

```python
%sql
-- Verificando quantas linhas tem na tabela:
SELECT COUNT(*) AS LINHAS
FROM workspace.olist_bronze.avaliacoes_bronze;
```

As correções a serem feitas são em relação ao tipo do dado e valores nulos.

Como a coluna `review_score` é a mais relevante para a análise, e as colunas `review_id` e `order_id` são essenciais para o relacionamento entre tabelas, não faz sentido manter valores nulos nessas colunas.

```python
%sql
-- Testando Query para criar a VIEW com segurança:
SELECT 
      review_id,
      order_id,
      CAST(review_score AS INT) AS review_score,
      CAST(review_creation_date AS TIMESTAMP) AS review_creation_date,
      CAST(review_answer_timestamp AS TIMESTAMP) AS review_answer_timestamp

FROM workspace.olist_bronze.avaliacoes_bronze
WHERE review_id IS NOT NULL AND review_id RLIKE '^[A-Za-z0-9]+$'
AND order_id IS NOT NULL AND order_id RLIKE '^[A-Za-z0-9]+$'
AND review_score IS NOT NULL;
```

```python
%sql
CREATE OR REPLACE VIEW workspace.olist_silver.avaliacoes_silver AS
SELECT 
    review_id,
    order_id,
    CAST(review_score AS INT)                   AS review_score,
    CAST(review_creation_date AS TIMESTAMP)     AS review_creation_date,
    CAST(review_answer_timestamp AS TIMESTAMP)  AS review_answer_timestamp
FROM 
    workspace.olist_bronze.avaliacoes_bronze
WHERE 
        review_id IS NOT NULL 
    AND review_id RLIKE '^[A-Za-z0-9]+$'
    AND order_id  IS NOT NULL 
    AND order_id  RLIKE '^[A-Za-z0-9]+$'
    AND review_score IS NOT NULL;
```
