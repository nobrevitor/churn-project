```python
%sql
-- ============================================================
-- CAMADA: SILVER
-- ENTIDADE: Pedidos
--
-- OBJETIVO:
-- Criar a tabela fato de pedidos na camada Silver, garantindo
-- integridade estrutural, padronização de tipos e vínculo
-- consistente entre pedidos e clientes, mantendo uma linha
-- por pedido.
--
-- FONTE:
-- workspace.olist_bronze.pedidos_bronze
--
-- DESTINO:
-- workspace.olist_silver.pedidos_silver
-- ============================================================

CREATE OR REPLACE VIEW workspace.olist_silver.pedidos_silver AS

SELECT
    p.order_id,
    p.customer_id,
    p.order_status,
    p.order_purchase_timestamp,
    p.order_approved_at,
    p.order_delivered_customer_date,
    p.order_estimated_delivery_date

FROM workspace.olist_bronze.pedidos_bronze p

WHERE p.order_id IS NOT NULL
  AND p.customer_id IS NOT NULL;
```
