```python
%sql
-- ============================================================
-- CAMADA: SILVER
-- ENTIDADE: Itens Agregados por Pedido
--
-- OBJETIVO:
-- Agregar os itens de cada pedido para a granularidade de
-- pedido, consolidando valores, quantidades e categorias
-- de produtos, garantindo consistência para análises
-- posteriores e modelos preditivos.
--
-- FONTE:
-- workspace.olist_bronze.itens_pedido_bronze
-- workspace.olist_bronze.produtos_bronze
-- workspace.olist_bronze.traducao_nome_produto_bronze
--
-- DESTINO:
-- workspace.olist_silver.itens_agregados_pedido_silver
-- ============================================================

CREATE SCHEMA IF NOT EXISTS workspace.olist_silver;

CREATE OR REPLACE VIEW workspace.olist_silver.itens_agregados_pedido_silver AS

SELECT
    i.order_id,
    COUNT(*) AS total_items,
    SUM(i.price) AS total_price,
    SUM(i.freight_value) AS total_freight,
    COUNT(DISTINCT p.product_category_name) AS distinct_product_categories

FROM workspace.olist_bronze.itens_pedido_bronze i

LEFT JOIN workspace.olist_bronze.produtos_bronze p
       ON i.product_id = p.product_id

GROUP BY i.order_id;
```
