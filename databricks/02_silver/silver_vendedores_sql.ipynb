## Tabela Vendedores:

### Objetivo:

Corrigir os erros identificados na camada bronze e criar a view vendedores_silver a partir da camada Bronze. 

### Fonte:

`workspace.olist_bronze.vendedores_bronze`

### Destino:

`workspace.olist_silver.vendedores_silver`

```python
%sql
-- Garantindo a exitência do SCHEMA
CREATE SCHEMA IF NOT EXISTS workspace.olist_silver;
```

```python
%sql
-- Verificando a existência da tabela:
SELECT * FROM workspace.olist_bronze.vendedores_bronze;
```

```python
%sql
-- Verificando quantas linhas tem na tabela:
SELECT COUNT(*) AS LINHAS
FROM workspace.olist_bronze.vendedores_bronze;
```

A coluna `seller_city` possui valores de cidade e estado concatenados, sendo necessário tratamento.

```python
%sql
SELECT
      seller_id,
      CAST(seller_zip_code_prefix AS INT) AS seller_zip_code_prefix,
      TRANSLATE(
          LOWER(TRIM(
              CASE 
                  WHEN seller_city LIKE '%@%' THEN NULL 
                  ELSE SPLIT(seller_city, '[,\\\\/]| -')[0]
              END
          )),
          'áàãâäéèêëíìîïóòõôöúùûüç',
          'aaaaaeeeeiiiiooooouuuuc'
      ) AS seller_city,
      seller_state
FROM workspace.olist_bronze.vendedores_bronze
```

```python
%sql
CREATE OR REPLACE VIEW workspace.olist_silver.vendedores_silver AS
SELECT
      seller_id,
      CAST(seller_zip_code_prefix AS INT) AS seller_zip_code_prefix,
      TRANSLATE(
          LOWER(TRIM(
              CASE 
                  WHEN seller_city LIKE '%@%' THEN NULL 
                  ELSE SPLIT(seller_city, '[,\\\\/]| -')[0]
              END
          )),
          'áàãâäéèêëíìîïóòõôöúùûüç',
          'aaaaaeeeeiiiiooooouuuuc'
      ) AS seller_city,
      seller_state
FROM workspace.olist_bronze.vendedores_bronze
```
