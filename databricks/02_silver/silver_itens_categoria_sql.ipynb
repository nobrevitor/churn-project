```python
%sql
-- ============================================================
-- CAMADA: SILVER
-- ENTIDADE: Itens por Categoria de Produto
--
-- OBJETIVO:
-- Criar uma tabela auxiliar contendo a relação entre pedidos,
-- itens e categorias de produtos, permitindo análises de
-- vendas e lucratividade por categoria no BI, sem adicionar
-- complexidade às tabelas factuais principais.
--
-- GRÃO:
-- 1 linha por item de pedido (order_id + order_item_id)
--
-- FONTES:
-- workspace.olist_bronze.itens_pedido_bronze
-- workspace.olist_bronze.produtos_bronze
--
-- DESTINO:
-- workspace.olist_silver.itens_categoria_silver
-- ============================================================
CREATE SCHEMA IF NOT EXISTS workspace.olist_silver;

CREATE OR REPLACE VIEW workspace.olist_silver.itens_categoria_silver AS

SELECT
    ip.order_id,
    ip.order_item_id,
    ip.product_id,
    pr.product_category_name

FROM workspace.olist_bronze.itens_pedido_bronze ip

LEFT JOIN workspace.olist_bronze.produtos_bronze pr
       ON ip.product_id = pr.product_id;
```
