```python
%sql
-- ============================================================
-- CAMADA: SILVER
-- ENTIDADE: Avaliações por Pedido
--
-- OBJETIVO:
-- Disponibilizar as avaliações dos clientes associadas aos
-- pedidos, mantendo a granularidade de um registro por pedido,
-- sem aplicação de regras de negócio ou interpretação dos
-- scores.
--
-- FONTE:
-- workspace.olist_bronze.avaliacoes_bronze
--
-- DESTINO:
-- workspace.olist_silver.avaliacoes_por_pedido_silver
-- ============================================================

CREATE SCHEMA IF NOT EXISTS workspace.olist_silver;

CREATE OR REPLACE VIEW workspace.olist_silver.avaliacoes_por_pedido_silver AS

SELECT
      order_id,
      review_score,
      review_creation_date,
      review_answer_timestamp
      
FROM workspace.olist_bronze.avaliacoes_bronze
WHERE order_id IS NOT NULL;
```
