## Tabela Geolocalização:

### Objetivo:

Corrigir os erros identificados na camada bronze e criar a view geolocalizacao_silver a partir da camada Bronze. 

### Fonte:

`workspace.olist_bronze.geolocalizacao_bronze`

### Destino:

`workspace.olist_silver.geolocalizacao_silver`

```python
%sql
-- Garantindo a exitência do SCHEMA
CREATE SCHEMA IF NOT EXISTS workspace.olist_silver;
```

```python
%sql
-- Verificando a existência da tabela:
SELECT * FROM workspace.olist_bronze.geolocalizacao_bronze;
```

```python
%sql
-- Verificando quantas linhas tem na tabela:
SELECT COUNT(*) AS LINHAS
FROM workspace.olist_bronze.geolocalizacao_bronze;
```

Serão aplicados dois tipos de correção:

- Alteração do tipo da coluna `geolocation_zip_code_prefix` de `BIGINT` para `INT`, a fim de otimizar o uso de memória.

- Padronização dos valores da coluna `geolocation_city`, uma vez que algumas cidades foram cadastradas com acentuação e outras sem.

```python
%sql
-- Testando Query para criar a VIEW com segurança:
SELECT
      CAST(geolocation_zip_code_prefix AS INT) AS geolocation_zip_code_prefix,
      geolocation_lat,
      geolocation_lng,

      TRANSLATE(
            LOWER(TRIM(geolocation_city)),
            'áàãâäéèêëíìîïóòõôöúùûüç',
            'aaaaaeeeeiiiiooooouuuuc'
            ) 
                                               AS geolocation_city,

      geolocation_state

FROM workspace.olist_bronze.geolocalizacao_bronze;
```

```python
%sql
CREATE OR REPLACE VIEW workspace.olist_silver.geolocalizacao_silver AS
  SELECT
        CAST(geolocation_zip_code_prefix AS INT) AS geolocation_zip_code_prefix,
        geolocation_lat,
        geolocation_lng,

        TRANSLATE(
              LOWER(TRIM(geolocation_city)),
              'áàãâäéèêëíìîïóòõôöúùûüç',
              'aaaaaeeeeiiiiooooouuuuc'
              ) 
                                                AS geolocation_city,

        geolocation_state

  FROM workspace.olist_bronze.geolocalizacao_bronze;
```
