```python
%sql
-- ============================================================
-- CAMADA: SILVER
-- ENTIDADE: Pagamentos Agregados por Pedido
--
-- OBJETIVO:
-- Consolidar os registros de pagamentos na granularidade de
-- pedido, evitando duplicidade de linhas e preparando os dados
-- para joins estruturais e futuras an√°lises financeiras.
--
-- FONTE:
-- workspace.olist_bronze.pagamentos_bronze
--
-- DESTINO:
-- workspace.olist_silver.pagamentos_agregados_silver
-- ============================================================

CREATE SCHEMA IF NOT EXISTS workspace.olist_silver;

CREATE OR REPLACE VIEW workspace.olist_silver.pagamentos_agregados_silver AS
SELECT
    order_id,
    SUM(payment_value) AS total_payment_value,
    MAX(payment_installments) AS max_installments,
    COUNT(DISTINCT payment_type) AS payment_types_count
FROM workspace.olist_bronze.pagamentos_bronze
GROUP BY order_id;
```
