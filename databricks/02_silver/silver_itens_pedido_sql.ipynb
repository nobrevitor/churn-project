## Tabela Itens por Pedido:

### Objetivo:

Corrigir os erros identificados na camada bronze e criar a view itens_pedido_silver a partir da camada Bronze. 

### Fonte:

`workspace.olist_bronze.itens_pedido_bronze`

### Destino:

`workspace.olist_silver.itens_pedido_silver`

```python
%sql
-- Garantindo a exitência do SCHEMA
CREATE SCHEMA IF NOT EXISTS workspace.olist_silver;
```

```python
%sql
-- Verificando a existência da tabela:
SELECT * FROM workspace.olist_bronze.itens_pedido_bronze;
```

```python
%sql
-- Verificando quantas linhas tem na tabela:
SELECT COUNT(*) AS LINHAS 
FROM workspace.olist_bronze.itens_pedido_bronze;
```

O único tratamento a ser aplicado é o tipo do dados da coluna `order_item_id`, mudando de `BIGINT` para `INT`.

```python
%sql
-- Testando Query para criar a VIEW com segurança:
SELECT
      order_id,
      CAST(order_item_id AS INT) AS order_item_id,
      product_id,
      seller_id,
      shipping_limit_date,
      price,
      freight_value
FROM workspace.olist_bronze.itens_pedido_bronze;
```

```python
%sql
CREATE OR REPLACE VIEW workspace.olist_silver.itens_pedido_silver AS
  SELECT
      order_id,
      CAST(order_item_id AS INT) AS order_item_id,
      product_id,
      seller_id,
      shipping_limit_date,
      price,
      freight_value
  FROM workspace.olist_bronze.itens_pedido_bronze;
```
