## Tabela Pagamentos:

### Objetivo:

Corrigir os erros identificados na camada bronze e criar a view pagamentos_silver a partir da camada Bronze. 

### Fonte:

`workspace.olist_bronze.pagamentos_bronze`

### Destino:

`workspace.olist_silver.pagamentos_silver`

```python
%sql
-- Garantindo a exitência do SCHEMA
CREATE SCHEMA IF NOT EXISTS workspace.olist_silver;
```

```python
%sql
-- Verificando a existência da tabela:
SELECT * FROM workspace.olist_bronze.pagamentos_bronze;
```

```python
%sql
-- Verificando quantas linhas tem na tabela:
SELECT COUNT(*) AS LINHAS
FROM workspace.olist_bronze.pagamentos_bronze;
```

```python
%sql
SELECT 
    COUNT(DISTINCT order_id)
FROM workspace.olist_bronze.pagamentos_bronze;
```

O único tratamentos aplicado será em relação ao tipo das colunas `payment_sequential` e `payment_intallments`, alterando de `BIGINT` para `INT`.

```python
%sql
-- Testando Query para criar a VIEW com segurança:
SELECT
      order_id,
      CAST(payment_sequential AS INT) AS payment_sequential,
      payment_type,
      CAST(payment_installments AS INT) AS payment_installments,
      payment_value
FROM workspace.olist_bronze.pagamentos_bronze;
```

```python
%sql
CREATE OR REPLACE VIEW workspace.olist_silver.pagamentos_silver AS
  SELECT
        order_id,
        CAST(payment_sequential AS INT) AS payment_sequential,
        payment_type,
        CAST(payment_installments AS INT) AS payment_installments,
        payment_value
        
  FROM workspace.olist_bronze.pagamentos_bronze;
```
