## Tabela Clientes:

### Objetivo:

Corrigir os erros identificados na camada bronze e criar a view clientes_silver a partir da camada Bronze. 

### Fonte:

`workspace.olist_bronze.clientes_bronze`

### Destino:

`workspace.olist_silver.clientes_silver`

```python
%sql
-- Garantindo a exitência do SCHEMA
CREATE SCHEMA IF NOT EXISTS workspace.olist_silver;
```

```python
%sql
-- Verificando a existência da tabela:
SELECT * FROM workspace.olist_bronze.clientes_bronze;
```

```python
%sql
-- Verificando quantas linhas tem na tabela:
SELECT COUNT(*) AS LINHAS
FROM workspace.olist_bronze.clientes_bronze;
```

A correção será feita na coluna `customer_zip_code_prefix` mudando o tipo `BIGINT` para `INT`, otimizando a utilização da memória. 

```python
%sql
-- Testando Query para criar a VIEW com segurança:
SELECT 
      customer_id,
      customer_unique_id,
      CAST(customer_zip_code_prefix AS INT) AS customer_zip_code_prefix,
      customer_city,
      customer_state
FROM workspace.olist_bronze.clientes_bronze;
```

```python
%sql
CREATE OR REPLACE VIEW workspace.olist_silver.clientes_silver AS
SELECT 
      customer_id,
      customer_unique_id,
      CAST(customer_zip_code_prefix AS INT) AS customer_zip_code_prefix,
      customer_city,
      customer_state
FROM workspace.olist_bronze.clientes_bronze;
```
