```python
%sql
-- ============================================================
-- CAMADA: SILVER
-- ENTIDADE: Clientes
--
-- OBJETIVO:
-- Criar a dimensão de clientes na camada Silver, consolidando
-- dados cadastrais e informações de localização, garantindo
-- uma linha por cliente e padronização estrutural para consumo
-- em BI e modelos analíticos.
--
-- FONTE:
-- workspace.olist_bronze.clientes_bronze
-- workspace.olist_bronze.geolocalizacao_bronze
--
-- DESTINO:
-- workspace.olist_silver.clientes_silver
-- ============================================================

CREATE SCHEMA IF NOT EXISTS workspace.olist_silver;

CREATE OR REPLACE VIEW workspace.olist_silver.clientes_silver AS
SELECT DISTINCT
    c.customer_id,
    c.customer_unique_id,
    c.customer_city,
    c.customer_state,
    g.geolocation_lat,
    g.geolocation_lng

FROM workspace.olist_bronze.clientes_bronze c
LEFT JOIN workspace.olist_bronze.geolocalizacao_bronze g
    ON c.customer_zip_code_prefix = g.geolocation_zip_code_prefix;
```
