### Tabela `avaliacoes`

```python
%sql
-- Volume total e registros com campos críticos nulos
SELECT
    COUNT(*) AS total_registros,
    SUM(CASE WHEN order_id IS NULL THEN 1 ELSE 0 END) AS order_id_nulos,
    SUM(CASE WHEN review_score IS NULL THEN 1 ELSE 0 END) AS review_score_nulos,
    SUM(CASE WHEN review_comment_title IS NULL THEN 1 ELSE 0 END) AS review_title_nulos,
    SUM(CASE WHEN review_comment_message IS NULL THEN 1 ELSE 0 END) AS review_message_nulos,
    SUM(CASE WHEN review_creation_date IS NULL THEN 1 ELSE 0 END) AS review_creation_date_nulos,
    SUM(CASE WHEN review_answer_timestamp IS NULL THEN 1 ELSE 0 END) AS review_answer_date_nulos


FROM workspace.default.avaliacoes;
```

```python
%sql
-- Identificar reviewID fora do padrão alfanumérico
SELECT review_id
FROM workspace.default.avaliacoes
WHERE review_id NOT RLIKE '^[a-zA-Z0-9]+$';
```

```python
%sql
-- Verificar cardinalidade entre pedidos e avaliações
SELECT
    order_id,
    COUNT(*) AS qtd_avaliacoes
FROM workspace.default.avaliacoes
GROUP BY order_id
HAVING COUNT(*) > 1;
```
