```python
%sql
-- ============================================================
-- CAMADA: GOLD (BI)
-- TABELA: dim_clientes
--
-- OBJETIVO:
-- Dimensão de clientes, usada para filtros, segmentações,
-- mapas e análises regionais no Power BI.
--
-- FONTE:
-- workspace.olist_silver.clientes_silver
--
-- DESTINO:
-- workspace.olist_gold.dim_clientes
-- ============================================================
CREATE SCHEMA IF NOT EXISTS workspace.olist_gold;

CREATE OR REPLACE VIEW workspace.olist_gold.dim_clientes AS

SELECT 
      customer_id,
      customer_city,
      customer_state

FROM workspace.olist_silver.clientes_silver
GROUP BY customer_id, customer_city, customer_state;
```
