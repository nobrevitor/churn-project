```python
%sql
-- ============================================================
-- CAMADA: GOLD
-- ENTIDADE: Itens por Categoria (BI)
--
-- OBJETIVO:
-- Disponibilizar a relação entre pedidos, itens e categorias
-- de produtos para consumo analítico no Power BI, permitindo
-- análises de vendas, volume e lucratividade por categoria.
--
-- ORIGEM:
-- workspace.olist_silver.itens_categoria_silver
--
-- DESTINO:
-- workspace.olist_gold.gold_itens_categoria
--
-- OBS:
-- View otimizada para BI, sem regras de negócio.
-- ============================================================
CREATE SCHEMA IF NOT EXISTS workspace.olist_gold;

CREATE OR REPLACE VIEW workspace.olist_gold.aux_itens_categoria AS
SELECT *
FROM workspace.olist_silver.itens_categoria_silver;
```
