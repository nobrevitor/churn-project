```sql
-- ============================================================
-- CAMADA: BRONZE
-- ENTIDADE: Geolocalizacao
--
-- OBJETIVO:
-- Criar a view geolocalizacao a partir da camada RAW,
-- aplicando apenas padronizações técnicas (sem regra de negócio).
--
-- FONTE:
-- workspace.default.geolocalizacao
--
-- DESTINO:
-- workspace.olist_bronze.geolocalizacao_bronze
-- ============================================================

CREATE SCHEMA IF NOT EXISTS workspace.olist_bronze;

CREATE OR REPLACE VIEW workspace.olist_bronze.geolocalizacao_bronze AS
SELECT
      CAST(geolocation_zip_code_prefix AS INT) AS geolocation_zip_code_prefix,
      CAST(geolocation_lat AS DOUBLE)          AS geolocation_lat,
      CAST(geolocation_lng AS DOUBLE)          AS geolocation_lng,
      LOWER(TRIM(geolocation_city))            AS geolocation_city,
      UPPER(TRIM(geolocation_state))           AS geolocation_state
FROM workspace.default.geolocalizacao;
```
