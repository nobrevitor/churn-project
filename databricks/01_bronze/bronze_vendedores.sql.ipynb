```python
%sql
-- ============================================================
-- CAMADA: BRONZE
-- ENTIDADE: Vendedores
--
-- OBJETIVO:
-- Criar a view vendedores a partir da camada RAW,
-- aplicando apenas padronizações técnicas (sem regra de negócio).
--
-- FONTE:
-- workspace.default.vendedores
--
-- DESTINO:
-- workspace.olist_bronze.vendedores_bronze
-- ============================================================

CREATE SCHEMA IF NOT EXISTS workspace.olist_bronze;

CREATE OR REPLACE VIEW workspace.olist_bronze.vendedores_bronze AS
SELECT
      CASE 
          WHEN seller_id RLIKE '^[a-zA-Z0-9_-]+$'
          THEN seller_id
          ELSE NULL
      END AS seller_id,

      CAST(seller_zip_code_prefix AS INT) AS seller_zip_code_prefix,
      LOWER(TRIM(seller_city))            AS seller_city,
      UPPER(TRIM(seller_state))           AS seller_state

FROM workspace.default.vendedores;
```
