```python
%sql
-- ============================================================
-- CAMADA: BRONZE
-- ENTIDADE: Produtos
--
-- OBJETIVO:
-- Criar a view produtos a partir da camada RAW,
-- aplicando apenas padronizações técnicas (sem regra de negócio).
--
-- FONTE:
-- workspace.default.produtos
--
-- DESTINO:
-- workspace.olist_bronze.produtos_bronze
-- ============================================================

CREATE SCHEMA IF NOT EXISTS workspace.olist_bronze;

CREATE OR REPLACE VIEW workspace.olist_bronze.traducao_nome_produto_bronze AS
SELECT
      CAST(product_category_name AS STRING)         AS product_category_name,
      CAST(product_category_name_english AS STRING) AS product_category_name_english

FROM workspace.default.traducao_nome_produto;
```
