```sql
-- ============================================================
-- CAMADA: BRONZE
-- ENTIDADE: Avaliacoes
--
-- OBJETIVO:
-- Criar a view bronze de avaliacoes a partir da camada RAW,
-- aplicando apenas padronizações técnicas (sem regra de negócio).
--
-- FONTE:
-- workspace.default.avaliacoes
--
-- DESTINO:
-- workspace.olist_bronze.avaliacoes_bronze
-- ============================================================
CREATE SCHEMA IF NOT EXISTS workspace.olist_bronze;

CREATE OR REPLACE VIEW workspace.olist_bronze.avaliacoes_bronze AS
SELECT
    CASE
        WHEN review_id RLIKE '^[a-zA-Z0-9_-]+$'
        THEN review_id
        ELSE NULL
    END AS review_id,

    CASE
        WHEN order_id RLIKE '^[a-zA-Z0-9_-]+$'
        THEN order_id
        ELSE NULL
    END AS order_id,

    CAST(review_score AS INT) AS review_score,
    review_comment_message,

    CAST(review_creation_date AS TIMESTAMP) AS review_creation_date,
    CAST(review_answer_timestamp AS TIMESTAMP) AS review_answer_timestamp

FROM workspace.default.avaliacoes;
```
