```sql
-- ============================================================
-- CAMADA: BRONZE
-- ENTIDADE: Itens do Pedido
--
-- OBJETIVO:
-- Criar a view itens_pedido a partir da camada RAW,
-- aplicando apenas padronizações técnicas (sem regra de negócio).
--
-- FONTE:
-- workspace.default.itens_pedido
--
-- DESTINO:
-- workspace.olist_bronze.itens_pedido_bronze
-- ============================================================

CREATE SCHEMA IF NOT EXISTS workspace.olist_bronze;

CREATE OR REPLACE VIEW workspace.olist_bronze.itens_pedido_bronze AS
SELECT
      CASE
        WHEN order_id RLIKE '^[a-zA-Z0-9_-]+$'
        THEN order_id
        ELSE NULL
      END AS order_id,
     
      CAST(order_item_id AS INT) AS order_item_id,

      CASE
        WHEN product_id RLIKE '^[a-zA-Z0-9_-]+$'
        THEN product_id
        ELSE NULL
      END AS product_id,  

      CASE
        WHEN seller_id RLIKE '^[a-zA-Z0-9_-]+$'
        THEN seller_id
        ELSE NULL
      END AS seller_id,      

      CAST(shipping_limit_date AS timestamp) AS shipping_limit_date,
      CAST(price AS DOUBLE)                  AS price,
      CAST(freight_value AS DOUBLE)          AS freight_value
      
FROM workspace.default.itens_pedido;
```
