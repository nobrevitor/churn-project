## Análise de Clientes com RFV

### Objetivo:

Este notebook tem como objetivo realizar uma análise exploratória e analítica dos dados da Olist, com foco em entender o comportamento de compra dos clientes a partir da base RFV previamente construída na camada de engenharia de dados.

A análise busca identificar padrões, perfis de clientes e insights acionáveis que possam apoiar decisões estratégicas de negócio e análises avançadas em ferramentas analíticas como Power BI.

### Sobre a Olist:

A Olist é uma plataforma brasileira de tecnologia que atua como um ecossistema completo para o varejo, com foco em democratizar o acesso ao e-commerce. Ela permite que lojistas vendam online com alta visibilidade, eficiência operacional e logística integrada, sem a necessidade de desenvolver soluções próprias. A Olist centraliza a gestão de vendas em múltiplos marketplaces, oferece ferramentas de ERP, soluções logísticas, serviços financeiros e plataforma para criação de loja virtual, apoiando o lojista em toda a jornada digital.
